```python
score = open('essentiality_pa.txt','r')     #change here 
mylist=list(score)
del mylist[0:3]
elist = []
gislist = []
for i in mylist:
    e,es,t,tgi= i.split('\t')
    g,gis,gbs,aad = tgi.split('|')
    elist.append(e)
    gislist.append(gis)

res = {}

for i in range(len(elist)):
    value = elist[i]
    key = gislist[i]
    res[key] = value

    Gi_Essential = []

for key, value in res.items():
    m = int(res[key])
    if m == 1:
        Gi_Essential.append(key)
print(Gi_Essential)
```

```python
import urllib.parse
import urllib.request

Uniprot_id = []
m = 1
for i in Gi_Essential:
    print(m)
    url = 'https://www.uniprot.org/uploadlists/'
    try:
        params = {
                    'from': 'P_GI',
                    'to': 'ACC',
                    'format': 'tab',
                    'query': i
        }
    
        data = urllib.parse.urlencode(params)
        data = data.encode('utf-8')
        req = urllib.request.Request(url, data)
        with urllib.request.urlopen(req) as f:
           response = f.read()
    
    except:
        continue
    else:
        
        res = response.decode('utf-8')
    
        res = res.replace("From\tTo\n", "")
        res = res.replace(i, "")
        res = res.replace("\t", "")
        res = res.replace("\n", "")
        Uniprot_id.append(res)
        m = m+1
print(Uniprot_id)
Uni = []

#Some of NCBI Gi may not have unique identifiers in Uniprot KB. Hence, they may be matched to Uniprac id. The following 
#Code will convert Uniprac ID into all possibilities of their UniprotKB. Our Uniprot ID of interest can be determined 
#in the next step 


for i in Uniprot_id:
    if len(i) == 13:
        irl = 'https://www.uniprot.org/uploadlists/'
        params = {
                'from': 'UPARC',
                'to': 'ACC',
                'format': 'tab',
                'query': i
        }
    
        data = urllib.parse.urlencode(params)
        data = data.encode('utf-8')
        req = urllib.request.Request(url, data)
        with urllib.request.urlopen(req) as f:
           response = f.read()
        res = response.decode('utf-8')
        
        res = res.replace("From\tTo\n", "")
        res = res.replace(i, "")
        res = res.replace("\t", "")
        res = res.replace("\n", ',')
        res = list(res.split(","))
        res = res[0:-1]
        Uni = Uni+res
    else:
        Uni.append(i)
print(Uni)
```

```python
from bioservices import KEGG
s = KEGG()
kegg_dict={}
for i in Uni:
    kegg_id = s.conv('pae','up:'+i)   #change here
    try:
       m = len(kegg_id.values())
    
    except:
        continue
    
    else:
        kegg_dict.update(kegg_id)
        
print(kegg_dict)
```
